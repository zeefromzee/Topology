# Shamir Secret Sharing Demo

This notebook demonstrates the implementation and use of Shamir's (k, n)-threshold secret sharing scheme.

## Overview

Shamir's secret sharing allows a secret to be divided into n shares such that any k shares can reconstruct the secret, but k-1 or fewer shares reveal no information about the secret.

## Topics Covered

1. Basic Shamir scheme implementation
2. Share generation and reconstruction
3. Security properties
4. Performance analysis
